```python
import geemap.foliumap as geemap
import geopandas as gpd
from shapely.geometry import Point

# Load and reproject
meta_data = gpd.read_file(meta_data_path).to_crs(epsg=4326)

# Create centroid GeoDataFrame separately
centroids = meta_data.geometry.centroid
centroid_gdf = gpd.GeoDataFrame(geometry=centroids, crs=meta_data.crs)

# Create map
Map = geemap.Map()
Map.add_gdf(meta_data, layer_name="Polygons")
Map.add_gdf(centroid_gdf, layer_name="Centroids", marker_type="circle", marker_radius=3, marker_color="red")

Map
```

```python
# Reproject to projected CRS (e.g., UTM zone 43N)
projected = meta_data.to_crs(epsg=32643)

# Compute centroids in projected CRS
projected["centroid"] = projected.geometry.centroid

# Reproject centroids back to geographic CRS
centroids = gpd.GeoSeries(projected["centroid"], crs=32643).to_crs(epsg=4326)

# Extract lon/lat
meta_data["lon"] = centroids.x
meta_data["lat"] = centroids.y

# Generate filename
meta_data["fn"] = meta_data["x"].astype(str) + "_" + meta_data["y"].astype(str) + ".tif"

# Save CSV
meta_data[["fn", "lon", "lat"]].to_csv(
    "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/india_bangladesh_pakisthan_images/index.csv",
    index=False
)
```
